```python
from numpy import *																								
																																		
N = 200																												
TOL = 0.000001																									
n = 3																															
a = [[5, 3, 2], [3, 4, -1], [-1, 1, -3]]								
b = [17, 8, -8]																									
																																		
x = zeros(n)																												
xp = zeros(n)																										
print("iter.       x1           x2           x3             Error")			
																																		
for k in range(1,N+1):																		
		print(k, end=" ")																						
		for i in range(n):																						
				xp[i] = x[i]																						
		for i in range(n):																						
				sum = 0																								
				for j in range(n):																					
						if j!=i:																							
								sum = sum + a[i][j] * xp[j]				
				x[i] = (b[i] - sum) / a[i][i]															
				print("      ", round(x[i],8), end=" ")														
																																		
		sum = 0																											
		for i in range(n):																							
				sum = sum + ((x[i] - xp[i]) * (x[i] - xp[i]))					
		err = sqrt(sum)																								
		print("    ", round(err,8))																							
		if err < TOL:																								
				break																								
																																		
print("The latest approximate solution vector is given:")					
for i in range(n):																							
		print(x[i], "\t", end=" ")																			
																																		
if err < TOL:																											
		print("\nThe desire accuracy is achieved; Solution is convergent.")		
else:																														
		print("\nThe number of iterations exceeded the maximum limit.")
```
